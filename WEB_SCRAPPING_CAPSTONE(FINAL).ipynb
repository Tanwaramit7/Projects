### Creating a Function for data extract and append data.

```python
def cars_extract(idx):
#     url_cars ='https://www.cars24.com/buy-used-car/?sort=P&storeCityId=2&pinId="
    cars24 = BeautifulSoup(requests.get(f"https://www.cars24.com/buy-used-car/?sort=P&storeCityId={city_id1[idx]}&pinId={city_id2[idx]}").content,'html5lib')
    r1 = cars24.find_all('div', attrs={'class':"col-4"})
    print('len r1',len(r1))
    car_name =[]
    km_driven=[]
    engine=[]
    emi=[]
    price=[]
    fuel = []
    for i in cars24.find_all('div', attrs={'class':"col-4"}):   #cars24.find_all('div', {'class':"col-4"})
        try:
            car_name.append(i.find('h2').text)
        except:
            car_name.append('n/a')
        try:
            km_driven.append(i.find('li').text)
        except:
            km_driven.append('n/a')
        try:
            engine.append(i.find('p').text)
        except:
            engine.append('n/a') 
        try:
            emi.append(i.find('strong').text)
        except:
            emi.append('n/a')
        try:
            price.append(i.find('div',{'class':"_7udZZ"}).text)
        except:
            price.append('n/a')
        try:
            fuel.append(i.find('ul',{'class':"bVR0c"}).text)#(strip=True, separator="|").split("|"))
        except:
            fuel.append('/')
    print(len(car_name),len(km_driven),len(engine),len(emi),len(price),len(fuel))
    car_city = pd.DataFrame({'car_name':car_name,'engine':engine,'km_driven':km_driven,'emi':emi,'price':price,'fuel':fuel})
    return car_city
```
